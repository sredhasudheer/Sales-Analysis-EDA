# Sales Analysis Project using Python

Objective:
To analyze the Superstore dataset and identify key insights related to sales, profit, regions, and product performance.

```python
import pandas as pd
df= pd.read_csv('SampleSuperstore.csv')
print(df.head())
print(df.columns)
print(df.isnull().sum())
print('Total Sales',df['Sales'].sum())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.groupby('Category')['Sales'].sum())
```

```python
print(df.groupby('Category')['Profit'].sum())
```

```python
print(df.groupby('Region')['Profit'].sum())                                      
```

```python
print(df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt

top_subcategories = (
    df.groupby('Sub-Category')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
plt.figure(figsize=(5,2))
top_subcategories.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

```python
print(df.groupby('Sub-Category')['Profit'].sum().sort_values())
```

```python
profit_subcategory=df.groupby('Sub-Category')['Profit'].sum().sort_values()
plt.figure(figsize=(4,2))
profit_subcategory.plot(kind='bar')
plt.title('profit by sub-category')
plt.xlabel('Sub-category')
plt.ylabel('Profit')
plt.show()
```

## Final Insights

• Technology generated the highest sales and profit.

• West region was the most profitable region.

• Central region had the lowest profit.

• Phones were among the highest-selling sub-categories.

• Tables generated the highest losses.

• Copiers generated the highest profit among sub-categories.
